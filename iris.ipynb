```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import tensorflow as tf
```

```python
iris=sns.load_dataset('iris')
x=iris.drop('species',axis=1)
y=iris['species']
print(y)
print(x.columns[0])
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
#Plots
summary_stats=x.describe()
print(summary_stats)
sns.set(style="ticks")
sns.pairplot(iris,hue='species')
plt.show()
```

```python
#Logistic Regression implementation
from sklearn.linear_model  import LogisticRegression
from sklearn import metrics
lr_model=LogisticRegression()
lr_model.fit(x_train,y_train)
y_pred=lr_model.predict(x_test)
#Accuracy
model_accuracy=metrics.accuracy_score(y_test,y_pred)

print(model_accuracy)
```

```python
#Prediction

import pandas as pd
flower_details=[]
for i in range(4):
  element=input(f"Enter {x.columns[i]} :")
  flower_details.append(element)
flower=np.array(flower_details,dtype=np.float64).reshape(-1,4)
preds=lr_model.predict(flower)
print("The given flower is {} with {}% accuracy".format(preds,(model_accuracy*100)))
#Plot
```
